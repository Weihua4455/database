## Scrape names and links
### for everyone except for local leaders

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
```

```python
import re

import pandas as pd
pd.set_option('display.max_colwidth', 500)
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("http://cpc.people.com.cn/GB/64162/394696/index.html")
```

```python
#Click on each tab, get html
tabs = driver.find_elements_by_tag_name("li")
```

```python
people_list = []

for tab in tabs[:-1]:
    tab_name = tab.text
    print(tab_name)
    tab.click()
    people = driver.find_elements_by_tag_name("td")
    
    for p in people:
        people_dict = {}
        people_dict["group"] = tab_name
        people_dict["name"] = p.text

        try:
            link = p.find_element_by_tag_name("a").get_attribute("href")
            people_dict["link"] = link
        except:
            people_dict["link"] = "NaN"
        people_list.append(people_dict)
```

### Great! Now I'll clean it in panda

```python
leader = pd.DataFrame(people_list)
leader.head()
```

```python
leader.shape
```

```python
leader["name"] = leader["name"].str.replace("\n（报道集）", "")
#leader["name"].value_counts()
```

```python
leader["name"] = leader["name"].str.replace(".*:", "")
leader["name"] = leader["name"].str.replace(".*：", "")
```

```python
leader["name"] = leader["name"].str.replace(".*\\n", "")
leader["name"] = leader["name"].str.replace("主席", "")
```

```python
#leader["name"].value_counts()
#leader["link"].value_counts()
```

### There are some duplicates, so I'm gonna drop the ones that have same url

```python
leader.drop_duplicates(subset=["link"], keep="first", inplace = True)
leader.shape
```
