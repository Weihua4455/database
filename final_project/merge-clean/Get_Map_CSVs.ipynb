```python
location_list = df.location.unique().tolist()
```

```python
df1 = pd.DataFrame()
```

```python
for place in location_list:
    df1 = df1.append(df[df.location == place].groupby("year").agg({"location":"first","location_lat": "first","location_lng": "first","name": ",".join}).reset_index())
```

```python
df1[df1.location == "辽宁"]
```

```python
def get_location_english(row):
    if row["location"] == "陕西":
        english = "Shaanxi"
    elif row["location"] == "香港":
        english = "Hong Kong"
    elif row["location"] == "西藏":
        english = "Xizang"
    elif row["location"] == "重庆":
        english = "Chongqing"
    else:
        english = pinyin.get(row["location"], format="strip").title()
    return english
```

```python
df1["location_english"] = df1.apply(get_location_english, axis=1)
```

```python
df1.location_english.unique()
```

```python
df1.to_csv("year_map.csv", index = False)
```
