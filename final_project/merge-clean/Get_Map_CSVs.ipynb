```python
location_list = df.location.unique().tolist()
```

```python
df1 = pd.DataFrame()
```

```python
for place in location_list:
    df1 = df1.append(df[df.location == place].groupby("year").agg({"location":"first","location_lat": "first","location_lng": "first","name": ",".join}).reset_index())
```

```python
df1[df1.location == "辽宁"]
```

```python
def get_location_english(row):
    if row["location"] == "陕西":
        english = "Shaanxi"
    elif row["location"] == "香港":
        english = "Hong Kong"
    elif row["location"] == "西藏":
        english = "Xizang"
    elif row["location"] == "重庆":
        english = "Chongqing"
    elif row["location"] == "内蒙古":
        english = "Inner Mongol"
    else:
        english = pinyin.get(row["location"], format="strip").title()
    return english
```

```python
df1["location_english"] = df1.apply(get_location_english, axis=1)
```

```python
df1.location_english.unique()
```

```python
df1.to_csv("year_map.csv", index = False)
```

### Now I want a file that contains information only on Xi Jinping

```python
df2 = pd.read_csv("for_map.csv")
```

```python
df2 = df2[df2.name == "习近平"]
```

```python
xi = pd.DataFrame()
```

```python
xi["content"] = "<b>" + df2["name"] + "</b>" + ": " + df2["content"]
```

```python
xi["year"] = df2["year"]
```

```python
def get_coordinates(row):
    coordinates = "[" + str(row["location_lat"]) + "," + str(row["location_lng"]) + "]"
    return coordinates
```

```python
xi["coordinates"] = df2.apply(get_coordinates, axis=1)
```

```python
xi.to_csv("xi_year_map.csv", index = False)
```
