```python
import requests
from bs4 import BeautifulSoup

import re
```

```python
my_url = "https://www.tribecafilm.com/filmguide?letter=b"
raw_html = requests.get(my_url)
soup_doc = BeautifulSoup(raw_html.content, "html.parser")
```

```python
print(soup_doc.prettify())
#soup_doc.title
```

The goal is pull out as much information as you can for each film from both pages films that start with B.

Ideally you want to end up with a list of dictionaries. But a list of lists is fine.

```python
#To start, see if you can get a list of all movie names
#If so, this is what you will loop throug to get all the information
movies = soup_doc.find_all("div", attrs = {"class" : "left"})
for movie in movies:
    try:
        movie_name = movie.find("h1", attrs = {"class" : "headerTitle"})
        print(movie_name.text)
    except:
        pass
```

```python
#first isolate the info for the first movie
movies = soup_doc.find_all("article")
movie_list = []

for movie in movies:
    left_text = movie.find("div", attrs = {"class" : "left"})
    movie_dic = {}
    try:
        #Name
        movie_name = left_text.find("h1", attrs = {"class" : "headerTitle"})
        movie_dic["Name"] = movie_name.string
        
        #Section
        movie_details = left_text.find("ul", attrs = {"class" : "filmDetails"})
        movie_section = movie_details.find_all("li")[0]
        movie_dic["Section"] = movie_section.string[9:]
        
        #Release Year
        release_year = movie_details.find_all("li")[1]
        movie_dic["Release_Year"] = release_year.string[-4:]
        
        #Runtime
        runtime = movie_details.find_all("li")[2]
        movie_dic["Runtime"] = runtime.string[9:]
        
        #DirectedBy
        director = movie_details.find_all("li")[3]
        if director.string != None:
            movie_dic["Director"] = director.string[13:]
        else:
            movie_dic["Director"] = "Data Missing"
        
        #Country
        country_list = []
        countries = movie_details.find_all("li")[-1].find_all("a")
        for country in countries:
            country_link = country["href"]
            country_name = re.findall("(?<=\=).*$", country_link)
            country_list.append(country_name)
        movie_dic["Country"] = country_list
         
        #ImageLink
        image_link = movie.img["src"]
        movie_dic["Image"] = image_link
        
        #Synopsis
        synopsis = movie.find("section", attrs = {"class" : "synopsis"}).text[:-10]
        movie_dic["Synopsis"] = synopsis
        
        #ReadMoreLink
        readmore = movie.find("a", attrs = {"class" : "readMore"})["href"]
        movie_dic["ReadMore"] = readmore

        #Genre
        genres = movie.find("section", attrs = {"class" : "genres"}).find_all("a")
        genre_list = []
        for genre in genres:
            genre_list.append(genre.string)
        movie_dic["Genre"] = genre_list
        
        #Tags
        tags = movie.find("div", attrs = {"class" : "headerTags"}).find_all("a")
        tag_list = []
        for tag in tags:
            tag_list.append(tag.string)
        movie_dic["Tag"] = tag_list
        
        #Add dictionary to the list
        movie_list.append(movie_dic)
    except:
        pass
print(movie_list)
```
